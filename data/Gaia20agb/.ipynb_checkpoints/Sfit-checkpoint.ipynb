## Color Test

```python
def fixed_data_binning(t, data, size=1):
    data_bin = []
    Gaia_arg = []
    for i in range(len(t)):
        arg1 = np.argwhere(t[i] - size <= data[:, 0])
        arg2 = np.argwhere(data[:, 0] < t[i] + size)
        temp = data[np.intersect1d(arg1, arg2), :]
        if len(temp) < 1:
            continue
        Gaia_arg.append(i)
        if len(temp) == 1:
            arg = 0
        else:
            arg = np.argwhere(
                abs(temp[:, 1] - temp[:, 1].mean()) <= 3 *
                np.std(temp[:, 1], ddof=1)).flatten()
        date_bin = temp[arg, 0].mean()
        mag_bin = (temp[arg, 1] / temp[arg, 2]**2).mean() / \
            (temp[arg, 2]**(-2)).mean()
        magerr_bin = ((data[arg, 2]**(-2)).mean())**(-0.5)
        data_bin.append([date_bin, mag_bin, magerr_bin])
        
    return np.array(data_bin), Gaia_arg
```

```python
def F(data, ZP=0):
    F = 10**(-0.4 * (data[:, 1]) - ZP)
    Ferr = abs(10**(0.4 * data[:, 2])-1) * F
    return F, Ferr
```

```python
def Color_test(A, G, size=10):
    Fa, Ferra = F(A)
    Fg, Ferrg = F(G)
    A_flux = np.array([A[:, 0], Fa, Ferra]).transpose()
    A_bin, arg = fixed_data_binning(G[:, 0], A_flux, size=size)
    Fa_bin, Ferra_bin = A_bin[:, 1], A_bin[:, 2]

    X = np.array([[(Fg[arg]**2).sum(), (Fg[arg]).sum()],
                  [(Fg[arg]).sum(), len(Fg[arg])]])
    Y = np.array([[(Fa_bin * Fg[arg]).sum()], [(Fa_bin).sum()]])
    A, B = np.matmul(np.linalg.inv(X), Y)
    chi2_reg = (((A * Fg[arg] + B - Fa_bin) / Ferra_bin)**2).sum()

    print(
        'Assuming no blending for telescope B, the estimated (relative) blending for telescope A is {:.2f}'
        .format((B / Fa)[0]))

    X = (Fg[arg]**2).sum()
    Y = (Fa_bin * Fg[arg]).sum()
    A0 = Y / X
    chi2_reg0 = (((A0 * Fg[arg] - Fa_bin) / Ferra_bin)**2).sum()

    f, ax = plt.subplots(3, 1, figsize=(6, 12), sharex=True)

    ax[0].errorbar(Fg[arg], Fa_bin, fmt='o', yerr=Ferra_bin)
    x = np.array([Fg[arg].min() / 1.1, Fg[arg].max() * 1.05])
    ax[0].plot(x, A * x + B, color='k', linestyle='--')
    ax[0].plot(x, A0 * x, color='k', linestyle=':')
    ax[0].set_ylabel('A flux', fontsize=25)

    ax[1].errorbar(Fg[arg], (Fa_bin - (A * Fg[arg] + B)) / Fa_bin * 100,
                   fmt='o',
                   yerr=Ferra_bin / Fa_bin * 100)
    ax[1].axhline(linestyle='--', color='k')
    ax[1].set_ylabel('Residual (b) [\%]', fontsize=25)
    print('Chi2 = {:.2f}'.format(chi2_reg))

    ax[2].errorbar(Fg[arg], (Fa_bin - (A0 * Fg[arg])) / Fa_bin * 100,
                   fmt='o',
                   yerr=Ferra_bin / Fa_bin * 100)
    ax[2].axhline(linestyle=':', color='k')
    ax[2].set_xlabel('B flux', fontsize=25)
    ax[2].set_ylabel('Residual [\%]', fontsize=25)
    print('Chi2 = {:.2f}'.format(chi2_reg0))

    for a in ax:
        a.tick_params(labelsize=20)
    f.tight_layout()
```

```python
Color_test(A=ASASSN, G=Gaia)
```

```python
LCOi_t = np.loadtxt('LCO-i.txt')
LCOr_t = np.loadtxt('LCO-r.txt')
LCOi_t[:, 0] -= 2450000
LCOr_t[:, 0] -= 2450000
Color_test(A=LCOi_t, G=data_binning(Gaia, 50), size=50)
```
