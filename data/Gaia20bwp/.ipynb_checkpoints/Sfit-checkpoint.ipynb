```python
def addfilter(dtab):
    """Add filter name as column in detection table by translating filterID
    
    This modifies the table in place.  If the 'filter' column already exists,
    the table is returned unchanged.
    """
    if 'filter' not in dtab.colnames:
        # the filterID value goes from 1 to 5 for grizy
        id2filter = np.array(list('grizy'))
        dtab['filter'] = id2filter[(dtab['filterID'] - 1).data]
    return dtab
```

```python
col = np.array(['#d7191c', '#fdae61', '#bbbbab', '#abd9e9', '#2c7bb6'])[::-1]
```

```python
objid = tab['objID'][0]
dconstraints = {'objID': objid}
dcolumns = (
    """objID,detectID,filterID,obsTime,ra,dec,psfFlux,psfFluxErr,psfMajorFWHM,psfMinorFWHM,
            psfQfPerfect,apFlux,apFluxErr,infoFlag,infoFlag2,infoFlag3"""
).split(',')
# strip blanks and weed out blank and commented-out values
dcolumns = [x.strip() for x in dcolumns]
dcolumns = [x for x in dcolumns if x and not x.startswith('#')]

dresults = ps1search(table='detection',
                     release='dr2',
                     columns=dcolumns,
                     **dconstraints)
dtab = addfilter(ascii.read(dresults))
dtab.sort('obsTime')

# convert flux in Jy to magnitudes
t = dtab['obsTime']
mag = -2.5 * np.log10(dtab['psfFlux']) + 8.90
magerr = -2.5 * np.log10(
    dtab['psfFlux']) + 2.5 * np.log10(dtab['psfFlux'] + dtab['psfFluxErr'])
xlim = np.array([t.min(), t.max()])
xlim = xlim + np.array([-1, 1]) * 0.02 * (xlim[1] - xlim[0])

f, ax = plt.subplots(figsize=(8, 8), sharex=True)
for i, filter in enumerate("grizy"):
    w = np.where(dtab['filter'] == filter)
    ax.errorbar(t[w],
                mag[w] - np.median(mag[w]),
                yerr=magerr[w],
                fmt='o',
                label=filter,
                color=col[i])
    x = np.array([t[w], mag[w], magerr[w]]).transpose()
    np.savetxt('Gaia20bwp_PS-{}.txt'.format(filter),
               x,
               fmt='%.3f',
               delimiter='  ')
    print(data_binning(x, 2000)[0])

ax.invert_yaxis()
ax.set_xlabel('Time [MJD]', fontsize=30)
ax.set_ylabel('Mag', fontsize=30)
ax.legend(prop={'size': 25})
plt.tight_layout()
plt.show()
```

```python
def ZTF_download(title, ra, dec, radius, bands='gri'):
    for b in bands:
        website = 'https://irsa.ipac.caltech.edu/cgi-bin/ZTF/nph_light_curves?POS=CIRCLE%20{:.5f}%20{:.5f}%20{:.5f}&BANDNAME={}&FORMAT=CSV'.format(
            ra, dec, radius / 3600, b)
        command = 'wget "{}" –O {}_ZTF-{}.csv'.format(website, title, b)
        #print(command)
        '''download = os.system(command)'''
        ZTFb = Table.read('{}_ZTF-{}.csv'.format(title, b), format='csv')
        ZTFb = np.array([ZTFb['hjd'], ZTFb['mag'],
                         ZTFb['magerr']]).transpose()
        ZTFb = data_binning(ZTFb, 1)
        np.savetxt('{}_ZTF-{}-PSF.txt'.format(title, b),
                   ZTFb,
                   fmt='%.6f',
                   delimiter='  ')


ZTF_download('Gaia20bwp', ra, dec, 1, bands='gr')
```

```python
def Add_Baseline(bl, source, telescope='LCO', bands='gri', binning=None):
    for b, band in zip(bl, bands):
        datafile = '{}_{}-{}-PSF.txt'.format(source, telescope, band)
        temp = np.loadtxt(datafile)
        if not binning==None:
            temp = data_binning(temp, binning)
        output = [b]
        if len(temp.shape) > 1:
            for t in temp:
                output.append(t)
        else:
            output.append(temp)
        file = '{}-{}.txt'.format(telescope, band)
        np.savetxt(file, output, fmt='%.3f', delimiter='  ')
```

```python
def colorterm(DIR,
              offset=[0, 0],
              flt='gr',
              xlim=[None, None],
              ylim=[None, None],
              g_r=0):
    from sklearn.linear_model import LinearRegression

    if len(offset) != len(flt):
        print('len(offset) != len(flt)')
        return
    f, ax = plt.subplots(len(flt),
                         1,
                         figsize=(10, 5 * len(flt)),
                         sharex=True,
                         sharey=True)
    ax = np.array([ax]).flatten()
    for k in range(len(flt)):
        t = Table.read(DIR + '{}p_gprp.csv'.format(flt[k]))
        t.remove_columns(['col1_2', 'col2_2'])
        t.rename_columns([
            'col1_1', 'col2_1', 'col3_1', 'col4_1', 'col3_2', 'col4_2', 'col5',
            'col6'
        ], [
            'RA', 'Dec', flt[k], flt[k] + 'err', 'g_PS', 'gerr_PS', 'r_PS',
            'rerr_PS'
        ])
        t[flt[k]] += offset[k]
        xerr = (t['gerr_PS']**2 + t['rerr_PS']**2)**(1 / 2)
        yerr = (t[flt[k] + 'err']**2 + t[flt[k] + 'err_PS']**2)**(1 / 2)
        x = t['g_PS'] - t['r_PS']
        y = t[flt[k]] - t[flt[k] + '_PS']
        arg1 = np.argwhere(ylim[0] > y)
        arg2 = np.argwhere(ylim[1] < y)
        arg = np.intersect1d(arg1, arg2)
        reg = LinearRegression().fit(x[arg].reshape(-1, 1), y[arg])
        print('{:.3e}, {:.3e}, {} - {}(PS) = {:.3f}'.format(
            reg.coef_[0], reg.intercept_, flt[k], flt[k],
            reg.coef_[0] * g_r + reg.intercept_))
        ax[k].errorbar(x,
                       y,
                       yerr=yerr,
                       fmt='o',
                       color='k',
                       ms=3,
                       elinewidth=.5)
        ax[k].set_ylabel('{} - {}(PS)'.format(flt[k], flt[k]), fontsize=25)
        X = np.array([0, 5])
        Y = reg.intercept_ + X * reg.coef_
        ax[k].plot(X, Y, color='grey')

    ax[-1].set_xlabel('g(PS) - r(PS)'.format(flt[k], flt[k]), fontsize=25)
    ax[-1].set_ylim(ylim)
    ax[-1].set_xlim(xlim)
    f.tight_layout()
```

```python
colorterm('std_ref_ZTF/', flt='gr', ylim=[0.2, -0.2], xlim=[0.5, 2.25], g_r=1.95)
```
